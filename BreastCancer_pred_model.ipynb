```python
# Data preprocessing

df.dropna(axis=1,inplace=True)
df
```

```python
# Resampling : under sampling

from sklearn.utils import resample

majority = df.loc[df['diagnosis']=='B']
minority = df.loc[df['diagnosis']=='M']

df_downSampled = resample(majority,
                          replace=False,
                          n_samples=len(minority),
                          random_state=42)


df_balanced = pd.concat([df_downSampled,minority])
```

```python
# Splitting into input and output data.

x = df_balanced.drop(columns=['id','diagnosis'])  # id is also removed because it is only used to identify uniquely not for prediction.
y = df_balanced['diagnosis']
```

```python
x.ndim
```

```python
y.ndim
```

```python
# train and test data split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# scaling

# Standard Scaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# x_res
x_train
```

```python
x_test
```

```python
# Model creation

# 1. KNN

knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(x_train,y_train)  #  The process of machine learing the pattern happens via the fit() here.
```

```python
y_pred_knn = knn.predict(x_test)
y_pred_knn
```

```python
# Performance Evaluation KNN

cm_knn = confusion_matrix(y_test,y_pred_knn)
cm_knn
```

```python
cmd_knn = ConfusionMatrixDisplay(cm_knn)
cmd_knn.plot()
```

```python
ac_knn = accuracy_score(y_test,y_pred_knn)
ac_knn*100
```

```python
# classification report for KNN Algorithm

report_knn = classification_report(y_test,y_pred_knn)
print(report_knn)
```

```python
# 2. Naive Bayes :

naive = GaussianNB()

naive.fit(x_train,y_train)
```

```python
y_pred_nb = naive.predict(x_test)
y_pred_nb
```

```python
# Performance Evaluation Navive Bayes
ac_nb = accuracy_score(y_test,y_pred_nb)
ac_nb*100
```

```python

cm_nb = confusion_matrix(y_test,y_pred_nb)
cm_nb
```

```python
cmd_nb = ConfusionMatrixDisplay(cm_nb)
cmd_nb.plot()
```

```python
# Classification report of Naive Bayes Algorithm.

report_nb = classification_report(y_test,y_pred_nb)
print(report_nb)
```

```python
# 3. Decision Tree classifier : ID3 Algorithm.

tree = DecisionTreeClassifier(random_state=42,criterion='entropy')

tree.fit(x_train,y_train)
```

```python
y_pred_tree = tree.predict(x_test)
y_pred_tree
```

```python
# Performance Evaluation Decision Tree

ac_tree = accuracy_score(y_test,y_pred_tree)
ac_tree*100
```

```python
cm_tree = confusion_matrix(y_test,y_pred_tree)
cm_tree
```

```python
cmd_tree = ConfusionMatrixDisplay(cm_tree)
cmd_tree.plot()
```

```python
# Classification report of Decision Tree Algorithm.

report_tree = classification_report(y_test,y_pred_tree)
print(report_tree)
```
